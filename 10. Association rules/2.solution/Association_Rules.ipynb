```python
import pandas as pd

# Load the data
data = pd.read_csv("/content/Online retail.csv")

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Keep only positive quantities and non-canceled invoices
# Check if the column name is 'Quantity' or 'quantity'
if 'Quantity' in data.columns:
  data = data[data['Quantity'] > 0]
else:
    # if 'Quantity' not found try 'quantity'
    if 'quantity' in data.columns:
        data = data[data['quantity'] > 0]
    else:
        print("Error:Neither 'quantity' nor 'Quantity' found in data columns. Please check the csv file.")

if 'InvoiceNo' in data.columns:
    data = data[~data['InvoiceNo'].str.startswith('C')]
else:
    # if 'InvoiceNo' not found try 'invoiceno'
    if 'invoiceno' in data.columns:
        data = data[~data['invoiceno'].str.startswith('C')]
    else:
        print("Error:Neither 'InvoiceNo' nor 'invoiceno' found in data columns. Please check the csv file.")

df = pd.get_dummies(data)
df.head()
```
